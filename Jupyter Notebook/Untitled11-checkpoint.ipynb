# Importing Basic Libraries

```python
# lets import the basic Libraries
import numpy as np
import pandas as pd

# for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# for Interactive Shells
from IPython.display import display

import pandas_profiling as pp

# setting up the chart size and background
plt.rcParams['figure.figsize'] = (16, 10)
plt.style.use('fivethirtyeight')
pd.set_option('display.max_column',23)
```

# Lets read the Dataset

```python
# lets read the dataset
data=pd.read_csv('ipl_match_ball_by_ball_data.csv')
data['season']=data['season'].replace(['2007/08'],'2008').replace(['2009/10'],'2010').replace(['2020/21'],'2020')
```

# Finding For missing values in each columns

```python
# check null values data1 
data.isnull().sum().sort_values(ascending=False)
```

```python
# lets remove unnecassary columns from the dataset

# Use the 'drop()' function to drop the unnecessary columns
data= data.drop(['legbyes',
                 'byes',
                 'penalty',
                 'noballs',
                 'wides',
                 'player_dismissed',
                 'other_wicket_type',
                 'other_player_dismissed'
                ],axis=1)
```

# Data Modification

```python
d=data.groupby(['season','striker'])['runs_off_bat'].sum().reset_index(name="Runs")

d['season']=d['season'].astype('int')

# Function write to find player all season Runs

def run(x):
    a = d[d['striker'] == x]
    y=a[['season','striker','Runs']]
    return y

# Function Call

df= run("V Kohli")

# Function after same Data Modification

d=df.sort_values(by=["season"])

player=d.groupby(['season','striker'])['Runs'].sum().reset_index(name ='All season runs')

player
```

```python
# Most Runs top 10 Player
data.groupby('striker')['runs_off_bat'].sum().sort_values(ascending =False).reset_index(name="Runs").head(10)
```

```python
# Most Wickets top 10 Player
data.groupby('bowler')['wicket_type'].count().sort_values(ascending =False).reset_index(name="Runs").head(10)
```

```python

# Orange Cap Holder All Seasons In Ipl

ee=data.groupby(['season','striker'])['runs_off_bat'].sum().reset_index(name="Runs")

ee['season']=ee['season'].astype('int')

ee.sort_values(by = ['season'])

dff=ee.groupby(['season','striker'])['Runs'].sum().reset_index(name ='All season runs')

dff = dff.sort_values(by = ["season","All season runs"],ascending =(True,False))

ss=dff.groupby('season')['All season runs'].head(1)

aa=pd.merge(ss, dff, left_index=True, right_index=True)

aa=aa.drop(['All season runs_x'], axis=1)

aa
```

```python

# Purple Cap Holder All Seasons Ipl

daa=data.groupby(['season','bowler'])['wicket_type'].count().reset_index(name="wicket")

daa['season']=daa['season'].astype('int')

daa.sort_values(by = ['season'])

dfff=daa.groupby(['season','bowler'])['wicket'].sum().reset_index(name ='All season wicket')

dfff = dfff.sort_values(by = ["season","All season wicket"],ascending =(True,False))

ssf=dfff.groupby('season')['All season wicket'].head(1)

aaf=pd.merge(ssf, dfff, left_index=True, right_index=True)

aaf=aaf.drop(['All season wicket_x'], axis=1)

aaf
```

```python

# Function Write Bat Man About Career all information  

def find(x):
    ds=data.groupby(['match_id','season','striker'])['runs_off_bat'].sum().reset_index(name="Runs")
    h=data.groupby('striker')['runs_off_bat'].sum().sort_values(ascending =False).reset_index(name="Runs").head(10)
    f=ds.loc[(ds['striker'] == x ) & (ds['Runs'] >= 50 ) & (ds['Runs'] <=99 )]
    m=ds.loc[ds['striker'] == x ]
    print("Innings:",m['Runs'].count())
    h=h.loc[h['striker'] == x]
    h = h['Runs']
    print("Aggregate:",h.sum())
    print("Average:",h.sum() / m['Runs'].count())
    p=ds.loc[(ds['striker'] == x ) & (ds['Runs'] > 20 )]
    c=ds.loc[(ds['striker'] == x ) & (ds['Runs'] > 99 )]
    print("Highest Score:",p['Runs'].sort_values(ascending=False).head(1).sum())
    print("Half Centuries:",f['Runs'].count())
    print("Centuries :",c['Runs'].count())
    s=ds.loc[ds['striker'] == x ]
    aa = data[data['striker'] == x]
    ya=aa[['season','striker','runs_off_bat']]
    ba= ya['runs_off_bat'].count()
    print("Balls Faced:",ba)  
    print("Scoring Rate:",h.sum() / ba*100)
    print("Top Scores in Innings:",f['Runs'].count() + c['Runs'].count())
    
   
   

find("V Kohli")    
```

```python

# Bowler ipl Career Wickets all season 

d=data.groupby(['season','bowler'])['wicket_type'].count().reset_index(name="top")

d['season']=d['season'].astype('int')


def run(x):
    
    a = d[d['bowler'] == x]
    
    y=a[['season','bowler','top']]
    
    return y

ds=run("DJ Bravo")

d=ds.sort_values(by=["season"])

df=d.groupby(['season','bowler'])['top'].sum().reset_index(name ='All season wicket_type')

df
```
